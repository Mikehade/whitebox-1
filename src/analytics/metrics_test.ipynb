```python
def format_feature_metrics(missing_count,non_missing_count,mean,minimum,maximum,sum,standard_deviation,variance):
    formated_metrics={'missing_count':missing_count,'non_missing_count':non_missing_count,'mean':mean,'minimum':minimum,'maximum':maximum,'sum':sum,'standard_deviation':standard_deviation,'variance':variance}

    return formated_metrics


def create_feature_metrics_pipeline(dataset):
    missing_count=dataset.isna().sum().to_dict()
    non_missing_count=dataset.notna().sum().to_dict()
    mean=dataset.mean(numeric_only=True).to_dict()
    minimum=dataset.min(numeric_only=True).to_dict()
    maximum=dataset.max(numeric_only=True).to_dict()
    sum=dataset.sum(numeric_only=True).to_dict()
    standard_deviation=dataset.std(numeric_only=True).to_dict()
    variance=dataset.var(numeric_only=True).to_dict()
    #count_unique_categorical=count_unique(dataset)

    
    return format_feature_metrics(missing_count,non_missing_count,mean,minimum,maximum,sum,standard_deviation,variance)
```

### Binary Classifier 

```python
dfcb=pd.read_csv("data/udemy_fin_adj.csv")
dfcb["is_paid"]=dfcb["is_paid"].map({True:0,False:1})
dfcb["discount_price__currency"]=dfcb["discount_price__currency"].map({'Dollar':0,'Euro':1,'Pound':3})
dfcb
```

```python
X=dfcb.drop(['is_paid'], axis=1)
y = dfcb.is_paid
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=1)
```

```python
clf = DecisionTreeClassifier()
clf = clf.fit(X_train,y_train)
y_pred = clf.predict(X_test)
```

```python
tn,fp,fn,tp = confusion_matrix(y_test, y_pred).ravel()
fp
```

```python
def format_evaluation_metrics_binary(accuracy, precision, recall, f1, tn, fp, fn, tp):
    formated_metrics_for_binary={'accuracy':accuracy,'precision':precision,'recall':recall,'f1':f1,'true_negative':tn,'false_positive':fp,'false_negative':fn,'true_positive':tp}

    return formated_metrics_for_binary

def create_binary_classification_evaluation_metrics_pipeline(test_set,prediction_set):
    accuracy=metrics.accuracy_score(test_set,prediction_set)
    precision=metrics.precision_score(test_set,prediction_set)
    recall=metrics.recall_score(test_set,prediction_set)
    f1=recall=metrics.f1_score(test_set,prediction_set)
    tn,fp,fn,tp = confusion_matrix(y_test, y_pred).ravel()

    return format_evaluation_metrics_binary(accuracy, precision, recall, f1, tn, fp, fn, tp)
    
```

```python
dfcm=pd.read_csv("data/udemy_fin_adj.csv")
dfcm["is_paid"]=dfcm["is_paid"].map({True:0,False:1})
dfcm["discount_price__currency"]=dfcm["discount_price__currency"].map({'Dollar':0,'Euro':1,'Pound':3})
dfcm
```

```python
X=dfcm.drop(['discount_price__currency'], axis=1)
y = dfcm.discount_price__currency
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=1)
```

```python
dtree_model = DecisionTreeClassifier().fit(X_train, y_train)
dtree_predictions = dtree_model.predict(X_test)
```

```python
cm=multilabel_confusion_matrix(y_test, dtree_predictions)
mult_dict={}
class_key=0
for i in cm:
    tn, fp, fn, tp = i.ravel()
    eval_dict= {'true_negative':tn,'false_positive':fp,'false_negative':fn,'true_positive':tp}
    mult_dict['class{}'.format(class_key)]=eval_dict
    class_key=class_key+1

```

```python
def confusion_for_multiclass(test_set,prediction_set):
    cm=multilabel_confusion_matrix(test_set,prediction_set)
    mult_dict={}
    class_key=0
    for i in cm:
        tn, fp, fn, tp = i.ravel()
        eval_dict= {'true_negative':tn,'false_positive':fp,'false_negative':fn,'true_positive':tp}
        mult_dict['class{}'.format(class_key)]=eval_dict
        class_key=class_key+1
    return mult_dict


def format_evaluation_metrics_multiple(accuracy, precision_statistics, recall_statistics, f1_statistics, conf_matrix):
    formated_metrics_for_multiple={'accuracy':accuracy,'precision_statistics':precision_statistics,'recall_statistics':recall_statistics,'f1_statistics':f1_statistics,'multiple_confusion_matrix':conf_matrix}

    return formated_metrics_for_multiple

def create_multiple_classification_evaluation_metrics_pipeline(test_set,prediction_set):
    accuracy=metrics.accuracy_score(test_set,prediction_set)
    micro_precision=metrics.precision_score(test_set,prediction_set, average='micro')
    macro_precision=metrics.precision_score(test_set,prediction_set, average='macro')
    weighted_precision=metrics.precision_score(test_set,prediction_set, average='weighted')
    precision_statistics={'micro':micro_precision,'macro':macro_precision,'weighted':weighted_precision}
    micro_recall=metrics.recall_score(test_set,prediction_set, average='micro')
    macro_recall=metrics.recall_score(test_set,prediction_set, average='macro')
    weighted_recall=metrics.recall_score(test_set,prediction_set, average='weighted')
    recall_statistics={'micro':micro_recall,'macro':macro_recall,'weighted':weighted_recall}
    micro_f1=recall=metrics.f1_score(test_set,prediction_set, average='micro')
    macro_f1=recall=metrics.f1_score(test_set,prediction_set, average='macro')
    weighted_f1=recall=metrics.f1_score(test_set,prediction_set, average='weighted')
    f1_statistics={'micro':micro_f1,'macro':macro_f1,'weighted':weighted_f1}
    conf_matrix = confusion_for_multiclass(y_test, y_pred)

    return format_evaluation_metrics_multiple(accuracy, precision_statistics, recall_statistics, f1_statistics, conf_matrix)
```
